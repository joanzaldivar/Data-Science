```python
#Top 5 states with the most counties
df_sub = df[["State", "County"]].groupby("State")["County"].nunique().nlargest(5)
df_sub
```

### Most Counties by State

```python
df_sub.sort_values().plot(kind="barh", y="State", figsize=(10, 5))
```

```python
#Top 5 states with the largest income per capita
df_sub = df[["State", "IncomePerCap"]].groupby("State")["IncomePerCap"].sum().nlargest(5)
df_sub
```

### Largest Income per Capita by State

```python
df_sub.sort_values().plot(kind="barh", y="State", figsize=(10, 5))
```

### Total Population by State

```python
#Get Total Population by State
df_sub = df[["State", "TotalPop"]].groupby('State')["TotalPop"].sum()
df_sub.head()
```

```python
# Visualize
# Sorted, horizontal bar
#df_sub.sort_values().plot(kind="barh", y="State", figsize=(10, 20))

# Sorted, vertical bar
df_sub.sort_values().plot(kind="bar", x="State", figsize=(20, 5))

# Unsorted, vertical bar
#df_sub.plot(kind="bar", x="State", figsize=(20, 5))
```

### Enhance Data Set
1. Calculate **Unemployed** as *Unemployment/100 x TotalPop* 
2. Calculate **NonWorkforce** as *TotalPop -(Unemployed + Employed)*

```python
df['Unemployed'] = ((df['Unemployment'] / 100)) * df['TotalPop']
```

```python
df['NonWorkforce'] = df['TotalPop'] - (df['Employed'] + df['Unemployed'])
#df.drop('Non-Workforce', axis=1, inplace=True)
```

### Correlation

```python
df.corr()
```

```python
#Correlation of income to poverty
df_sub = pd.pivot_table(df, values=["Income","Poverty"], index="State", aggfunc=np.mean)
df_sub.plot(kind="scatter", y="Income", x="Poverty")
```

### Poverty by State

```python
df_sub["Poverty"].sort_values().plot(kind="bar", y="Poverty", x=df_sub.index, figsize=(20,5))
```

### Plotting Poverty and Income

```python
#plt.plot(df_sub[df_sub.index], df_sub["Poverty"], df_sub[df_sub.index], df_sub["Income"])

df_sub = df.groupby("State", as_index=False).mean()
```

```python
df_sub.columns #State is part of the columns
```

#### Separate Plots by State

```python
x = df_sub["State"]
y1 = df_sub["Poverty"]
y2 = df_sub["Income"]

plt.figure(figsize=(20,5))
plt.plot(x,y1)
plt.legend(['y1'])
plt.savefig('two-scales-2.png')

plt.figure(figsize=(20,5))
plt.plot(x,y2)
plt.legend(['y2'])
plt.savefig('two-scales-3.png')
```

#### Combined Plots by State

```python
plt.figure(figsize=(20,5))
plt.plot(x, y1, x, y2)
plt.legend(['Poverty', 'Income'])
plt.savefig('two-scales-4.png')

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(111)
ax1.plot(x, y1)
ax1.set_ylabel('Poverty')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'r-')
ax2.set_ylabel('Income', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.savefig('two-scales-5.png')
```

```python
plt.figure(figsize=(20,5))
f, axes = plt.subplots(2, 1)
axes[0].plot(x, y1)
axes[0].set_ylabel('Poverty')

axes[1].plot(x, y2)
axes[1].set_ylabel('Income')
plt.savefig('two-scales-6.png')
```
